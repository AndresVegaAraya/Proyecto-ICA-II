```python
import pandas as pd

# 1. Define las rutas de tu archivo de origen y el de destino
ruta_excel = 'OT.xlsx'
ruta_parquet = 'OT.parquet'

# 2. Lee el archivo Excel y cárgalo en un DataFrame
df = pd.read_excel(ruta_excel, engine='openpyxl')

# 3. Solución (Opción B): Encuentra todas las columnas de tipo 'object' 
# (texto mezclado con números/nulos) y conviértelas estrictamente a texto (string)
columnas_mixtas = df.select_dtypes(include=['object']).columns
df[columnas_mixtas] = df[columnas_mixtas].astype(str)

# 4. Guarda el DataFrame en formato Parquet
df.to_parquet(ruta_parquet, engine='pyarrow')

print("¡Archivo convertido a Parquet con éxito sin errores de formato!")
```
